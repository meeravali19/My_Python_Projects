```python
import pyttsx3
from PyPDF2 import PdfReader

# Initialize the text-to-speech engine (assuming you are using pyttsx3)
speaker = pyttsx3.init()

# Open the PDF file
pdf_file_path = 'Problem Statement.pdf'

# Create a PdfReader object to read the PDF file
pdf_reader = PdfReader(pdf_file_path)

# Iterate over each page
for page_number in range(len(pdf_reader.pages)):
    # Get the specific page
    page = pdf_reader.pages[page_number]
    
    # Extract text from the page
    text = page.extract_text()
    
    # Print the extracted text (optional)
    print(text)
    
    # Speak out the extracted text using the text-to-speech engine
    speaker.say(text)

# Wait for the speech to complete
speaker.runAndWait()
```

```python
import pyttsx3
from PyPDF2 import PdfReader
import keyboard

# Initialize the text-to-speech engine (assuming you are using pyttsx3)
speaker = pyttsx3.init()

# Function to speak the text and handle keyboard input
def speak_text(text):
    speaker.say(text)
    speaker.runAndWait()

    # Listen for keyboard input to control speech
    while True:
        if keyboard.is_pressed('s'):  # Press 's' to stop speaking
            speaker.stop()
            break
        elif keyboard.is_pressed('p'):  # Press 'p' to pause/resume speaking
            if speaker._inLoop:
                speaker.pause()
            else:
                speaker.resume()

# Open the PDF file
pdf_file_path = 'Problem Statement.pdf'

# Create a PdfReader object to read the PDF file
pdf_reader = PdfReader(pdf_file_path)

# Iterate over each page
for page_number in range(len(pdf_reader.pages)):
    # Get the specific page
    page = pdf_reader.pages[page_number]
    
    # Extract text from the page
    text = page.extract_text()
    
    # Print the extracted text (optional)
    print(text)
    
    # Speak out the extracted text using the text-to-speech engine with control
    speak_text(text)

    # Check for keyboard input to continue or stop
    if page_number < len(pdf_reader.pages) - 1:
        print("\nPress 's' to stop, 'p' to pause/resume, or any other key to continue to the next page...")
        # Wait for key press before continuing to the next page
        keyboard.wait()
```
