```python
import cv2

def QRreader(image_path):
    # Read the input image
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Could not read the image.")
        return None

    # Create a QRCodeDetector object
    detector = cv2.QRCodeDetector()

    # Detect and decode QR code
    data, bbox, _ = detector.detectAndDecode(img)

    if data:
        
        return data
    else:
        
        return None

if __name__ == "__main__":
    # Get the path to the image containing the QR code
    image_path = input("Enter the path to the image containing the QR code: ")

    # Call the QRreader function
    decoded_data = QRreader(image_path)

    if decoded_data:
        print("Decoded QR code data:", decoded_data)
    else:
        print("Failed to decode QR code from the provided image.")
```
